### filling na columns

```python
df.shape[0]
```

```python
check_na_columns(df)
```

```python
cols_to_replace = ["fcr_flag","fcr_amount","product_qty","payment_type","fcr_rate_of_exchange","product_amt"]
```

```python
for col in cols_to_replace:
    df[col] = df[col].fillna(0)
check_na_columns(df)
```

```python
df["adj_orgn_tran_dt"] = df["adj_orgn_tran_dt"].fillna("No Date")
df["invoice_nbr"] = df["invoice_nbr"].fillna("No Invoice Number")
check_na_columns(df)
```

```python
df.head()
```

```python
df["transaction_date"] = pd.to_datetime(df["transaction_date"], errors="coerce")
df["adj_orgn_tran_dt"] = pd.to_datetime(df["adj_orgn_tran_dt"], errors="coerce")
df["posting_date"] = pd.to_datetime(df["posting_date"], errors="coerce")
df.dtypes
```
